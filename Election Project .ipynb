```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data4 = pd.read_csv('LokSabha2019.csv')
```

```python
# 1. Age Wise Distribution Of The Politicians
a = data4['Age'] 
plt.title('Age Distribution Of Politicians')
plt.xlabel('Range Of Age Of Candidates')
plt.ylabel('No. Of People')
plt.hist(a , edgecolor = "black")
plt.tight_layout()
plt.style.use('fivethirtyeight')
plt.legend()
```

```python
# 2. Distribution Between Male and Female On The Basis Of Political Party 
b = data4['Gender'] == 'F'
c = data4['Winner'] == 1.0
d = data4[b & c]
g = d['Party'].value_counts().rename_axis('Party').reset_index(name='Candidates').head()
x_co = g['Party']
y_co = g['Candidates']

k = sns.barplot(x_co  , y_co )
k.set(ylabel = "Female Leaders", title ='Female Leaders In Individual Parties (won)')
```

```python
b1 = data4['Gender'] == 'M'
c1= data4['Winner'] == 1.0
d1 = data4[b1 & c1]
g1 = d1['Party'].value_counts().rename_axis('Party').reset_index(name='Candidates').head()
x_co = g1['Party']
y_co = g1['Candidates']
g = sns.barplot(x_co  , y_co )
g.set(ylabel = "Male Leaders", title ='Males Leaders In Individual Parties (won)')
```

```python
# 3. Most Corroupt Politicians
a = data4['Total Assets']  > 3500000000
g = data4[a]
p = g['Candidate'].value_counts().rename_axis('Candidates').reset_index(name='values')
k = []
for i in range(0,5):
    l = p['Candidates'][i][0:7]
    k.append(l)
y = g['Total Assets']
plt.title('Assests By An Individual')
plt.xlabel('Names Of Leaders')
plt.ylabel('Assests They Have')
plt.tight_layout()
plt.style.use('ggplot')
plt.legend()
plt.plot(k , y , marker='o')
```

```python
a = data4['Party'].value_counts()
labels = ['IND' , 'BJP' , 'INC' ,'BSP']
fig, ax = plt.subplots()
plt.title('Leaders Ratio Between Top 5 Parties')
plt.tight_layout()
plt.style.use('ggplot')
plt.legend()
plt.pie(a.head(4) , labels = labels , autopct='%.0f%%' , explode=[0.03 , 0.03 , 0.03 ,0.03])
plt.show()
```

```python
a = data4['Criminal Cases'] >= 1
b = data4[a]
c = b['Party'].value_counts()
plt.title('Leaders With Criminal Cases')
plt.xlabel('Parties')
plt.ylabel('No. Of Leaders')
plt.tight_layout()
plt.style.use('seaborn-dark-palette')
plt.legend()
plt.plot(c.head() , marker='o')
```

```python
a = data4['Gender'] == 'M'
b = data4[a]
y = b['Criminal Cases']
x = b['Age']
plt.scatter(x , y , label='Male')

a1 = data4['Gender'] == 'F'
b1 = data4[a1]
y1 = b1['Criminal Cases']
x1 = b1['Age']
plt.scatter(x1 , y1 , label='Female' )

plt.title('Criminal Cases Against Male Vs Female')
plt.xlabel('Age')
plt.ylabel('Criminal Cases')
plt.tight_layout()
plt.style.use('ggplot')
plt.legend()

```

```python
a = data4['Criminal Cases'] == 0
b = data4['Liabilities'] == 0
c = data4['Education'] != 'Illiterate'
d = data4[a & b & c]
e = d['Party'].value_counts()

plt.title('Leaders With 0 Criminal Cases and Liabilities and Is Literate')
plt.xlabel('Party')
plt.ylabel('No. Of Candidates')
plt.tight_layout()
plt.style.use('ggplot')
plt.legend()
plt.plot(e.head() ,  marker='o')
```

```python
a = data4['Liabilities'] >= 90000000
b = data4['Winner'] == 1.0
c = data4[a & b]
x = c['Age']
y  = c['Liabilities']
plt.title('Leader Who is a Liabilities For The Country And Has Won The Election')
plt.xlabel('Party')
plt.ylabel('Liability (In Rupree)')
plt.tight_layout()
plt.style.use('ggplot')
plt.legend()
plt.scatter(x , y)
```
