```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data4 = pd.read_csv('LokSabha2019.csv')
```

```python
# 1. Age Wise Distribution Of The Politicians
a = data4['Age'] 
plt.title('Age Distribution Of Politicians')
plt.xlabel('Range Of Age Of Candidates')
plt.ylabel('No. Of People')
plt.hist(a , edgecolor = "black")
plt.tight_layout()
plt.style.use('fivethirtyeight')
# plt.show()
plt.legend()
```

```python
# 2. Distribution Between Male and Female On The Basis Of Political Party 
b = data4['Gender'] == 'F'
c = data4['Winner'] == 1.0
d = data4[b & c]
g = d['Party'].value_counts()


b1 = data4['Gender'] == 'M'
c1= data4['Winner'] == 1.0
d1 = data4[b1 & c1]
g1 = d1['Party'].value_counts()
```

```python
# 3. Most Corroupt Politicians
a = data4['Total Assets']  > 3500000000
g = data4[a]
x = g['Candidate']
y = g['Total Assets']
sns.catplot(x , y , data = g)
```
